###**Loading the Dataset**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic_df = pd.read_csv("Titanic.csv")
titanic_df
```

###**Histogram of Ages**

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_df, x='Age', bins=20, kde=True, color='skyblue')
plt.title('Distribution of Ages on Titanic')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

###**Histogram of Gender**

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic_df, x='Sex', palette='pastel')
plt.title('Distribution of Gender on Titanic')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()
```
