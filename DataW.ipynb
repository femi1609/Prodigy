```python
s.boxplot(column="Insulin")
#Here boxplot is plotted to detect if there is any outlier in the particular column of the dataset
```

```python
import numpy as np
np.where(s["Insulin"]>300)
#Looking at the above plot, we can find where or at what index those outliers are lying using numpy
```

```python
q1=s["Insulin"].quantile(0.25)
q1
#First Quartile
```

```python
q3=s["Insulin"].quantile(0.75)
q3
#Third Quartile
```

```python
iqr=q3-q1
iqr
#Inter quartile range between the third quartile and first quartile
```

```python
l=q1-0.5*iqr
l
#Lower found using subtracting quartile deviation from first quartile
```

```python
u=q3-0.5*iqr
u
```

```python
upper_array = np.where(s['Insulin']>=u)[0]
lower_array = np.where(s['Insulin']<=l)[0]
upper_array
```

```python
lower_array
```

```python
# Removing the outliers
s.drop(index=upper_array, inplace=True)
s.drop(index=lower_array, inplace=True)
 
# Print the new shape of the DataFrame
print("New Shape: ",s.shape)
```
