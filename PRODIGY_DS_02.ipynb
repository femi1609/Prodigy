###**Loading the Dataset**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic_df = pd.read_csv("Titanic.csv")
titanic_df
```

```python
print("First few rows of the dataset:")
titanic_df.head()
```

```python
print("\nMissing values in the dataset:")
titanic_df.isnull().sum()
```

```python
median_age = titanic_df['Age'].median()
titanic_df['Age'].fillna(median_age, inplace=True)
```

```python
titanic_df
```

```python
mode_embarked = titanic_df['Embarked'].mode()[0]
titanic_df['Embarked'].fillna(mode_embarked, inplace=True)
titanic_df
```

```python
# Dropping the Cabin column due to a large number of missing values
titanic_df.drop(columns=['Cabin'], inplace=True)
```

```python
print("\nData types of each column:")
print(titanic_df.dtypes)
```

```python
print("\nSummary statistics for numerical columns:")
titanic_df.describe()
```

```python
print("\nSummary statistics for categorical columns:")
print(titanic_df.describe(include=['object']))
```

```python
# Distribution of Survival
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic_df, x='Survived', palette='Set1')
plt.title('Distribution of Survival')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of Pclass
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic_df, x='Pclass', palette='deep')
plt.title('Distribution of Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation matrix
import numpy as np
plt.figure(figsize=(10, 6))
numeric_cols = titanic_df.select_dtypes(include=[np.number]).columns
correlation_matrix = titanic_df[numeric_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
